<br>Дополнить класс таким образом, чтобы выполнялся следующий код:<br>
<code>frac = Fraction(7, 2)</code><br>
<code>print(-frac)</code> # выводит -7/2<br>
<code>print(~frac)</code> # выводит 2/7<br>
<code>print(frac**2)</code> # выводит 49/4<br>
<code>print(float(frac))</code> # выводит 3.5<br>
<code>print(int(frac))</code> # выводит 3 

```python
class Fraction(object):
    def __init__(self, num, den):
        self.__num = num
        self.__den = den
        self.reduce()
    def __str__(self):
        return "%d/%d" % (self.__num, self.__den)
    def reduce(self):
        g = Fraction.gcd(self.__num, self.__den)
        self.__num /= g
        self.__den /= g
    @staticmethod
    def gcd(n, m):
        if m == 0:
            return n
        else:
            return Fraction.gcd(m, n % m)


    def __neg__(self):
        return Fraction(-self.__num, self.__den)

    def __invert__(self):
        return Fraction(self.__den, self.__num)

    def __pow__(self, number):
        return Fraction(self.__num ** number, self.__den ** number)
    
    def __float__(self):
        return round(self.__num / self.__den, 2)

    def __int__(self):
        return int(self.__float__())

        

frac = Fraction(7, 2)
print("frac = ", frac)
print("-frac = ", -frac)
print("~frac = ", ~frac)
print("frac**2 = ", frac**2)
print("float(frac) = ", float(frac))
print("int(frac) = ", int(frac))
```

```python
import re
from abc import abstractmethod

class Taggable:

    @abstractmethod
    def tag(self):
        pass

class Book(Taggable): 
   
    def __init__(self, author, name):  
        if (name == "" or author ==""):
            raise ValueError
        else:
            self.__name = name
            self.__author = author
        self.code = 0
        
    def tag(self): 
        return re.compile('[A-Z][a-z]+').findall(self.__name)  

    def __str__(self): 
        return "[%s] %s '%s'" % (str(self.code), self.__author, self.__name)



class Library(object):  
    code = 1  

    def __init__(self, address, number): 
        
        self.__address = address  
        self.__number = number 
        self.__books = []  

    def __iadd__(self, book):  
        book.code = Library.code
        Library.code += 1
        self.__books.append(book)  
        return self 

    def __iter__(self): 
        return iter(self.__books) 

library = Library ("83001, г. Донецк, ул. Артема, 84", 1)
library += Book ("Timofei Neshariy", "I love Java")
for book in library:
    print(book)
    print (book.tag())
```

```python
import wx
import os
import re
import datetime
fileName = ""

def main_func(path):
    global fileName
    fileName = path
    with open(os.path.join(os.getcwd(), fileName), 'r') as f:
        temp = f.readlines()

    list = []
    for i, line in enumerate(temp):
        elem: list = re.findall(r"[;:]-*(\(+|\)+|\[+|\]+)", line), i + 1
        if elem: list.append(elem)
    res = []
    for elem, i in list:
        for x in elem:
            res.append(f'Строка {i}, позиция {temp[i - 1].find(x)} : найдено {x}')
   
    return res


def print_to_lb(list_, list_box):
    
    list_box.Append('Файл ' + fileName + ' был обработан ' + datetime.datetime.now().strftime('%d.%m.%Y %H:%M:%S') + ':')
    for item in list_:
        list_box.Append(item)


def check_log_file(app):
    if not os.path.isfile('./script18.log'):
        dlg = wx.MessageBox('Файл лога не найден. Файл будет создан автоматически', 'Внимание!', wx.OK | wx.ICON_EXCLAMATION, app)
        with open('./script18.log', 'w') as f:
            print('File Created')


class MyPanel(wx.Panel):
    def __init__(self, parent):
        super().__init__(parent)

        vbox = wx.BoxSizer()
        self.list_box = wx.ListBox(self, wx.ID_ANY, style=wx.LB_SINGLE)
        vbox.Add(self.list_box, 1, wx.ALL | wx.EXPAND, 5)
        self.SetSizer(vbox)


class MyMenu(wx.MenuBar):
    def __init__(self):
        super().__init__()

        file_menu = wx.Menu()
        log_menu = wx.Menu()

        open_b = file_menu.Append(wx.ID_ANY, 'Открыть')
        export_b = log_menu.Append(wx.ID_ANY, 'Экспорт')
        add_b = log_menu.Append(wx.ID_ANY, 'Добавить в лог')
        look_b = log_menu.Append(wx.ID_ANY, 'Просмотр')

        self.Append(file_menu, 'Файл')
        self.Append(log_menu, 'Лог')

        self.Bind(wx.EVT_MENU, self.open_file, open_b)
        self.Bind(wx.EVT_MENU, self.add_to_log, add_b)
        self.Bind(wx.EVT_MENU, self.export_to_file, export_b)
        self.Bind(wx.EVT_MENU, self.look_log, look_b)

    def look_log(self, e):
        dlg = wx.MessageDialog(frame, 'Вы действительно хотите открыть лог? Данные последних поисков будут потеряны!', 'Внимание!', wx.YES_NO | wx.NO_DEFAULT | wx.ICON_EXCLAMATION)
        res = dlg.ShowModal()
        if res == wx.ID_YES:
            frame.panel.list_box.Clear()
            with open('./script18.log', 'r') as f:
                lines = f.readlines()
            for line in lines:
                frame.panel.list_box.Append(line[:len(line) - 1])
            frame.status_bar.SetStatusText('Открыт лог')
        if res == wx.ID_NO:
            return 'NO'

    def add_to_log(self, e):
        with open('./script18.log', 'a') as f:
            for line in frame.panel.list_box.GetStrings():
                f.write(f'{line}\n')

    def open_file(self, e):
        with wx.FileDialog(self, 'Открыть файл…', style=wx.FD_OPEN) as fd:
            if fd.ShowModal() == wx.ID_CANCEL:
                return 'cancel'
            path_name = fd.GetPath()
            text = main_func(path_name)
            frame.panel.list_box.Clear()
            print_to_lb(text, frame.panel.list_box)
            frame.status_bar.SetStatusText(f'Обработан файл {os.path.dirname(path_name)}')
            frame.status_bar.SetStatusText('{0:,} байт'.format(os.path.getsize(path_name)).replace(',', ' '), 1)

    def export_to_file(self, e):
        with wx.FileDialog(self, 'Экспорт в…', style=wx.FD_OPEN) as fd:
            if fd.ShowModal() == wx.ID_CANCEL:
                return 'cancel'
            path_name = fd.GetPath()
            with open(path_name, 'w') as f:
                for line in frame.panel.list_box.GetStrings():
                    f.write(f'{line}\n')


class MyFrame(wx.Frame):
    def __init__(self, parent = None):
        super().__init__(parent)
        self.SetTitle('Искатель строк')
        self.Center()
        self.SetSize((600,400))
        path = os.getcwd()
        ico = wx.Icon( path + '\\' + 'string.ico', wx.BITMAP_TYPE_ICO)
        self.SetIcon(ico)
        self.SetMaxSize((600, 400))
        self.SetMinSize((600, 400))
        menubar = MyMenu()
        self.SetMenuBar(menubar)
        self.panel = MyPanel(self)
        self.status_bar = self.CreateStatusBar(2)
        self.status_bar.SetStatusWidths([-6, -4])
        self.Show()
        check_log_file(self)


app = wx.App()
frame = MyFrame()
app.MainLoop()
```

– замена всех цифр в строке на знак «*»;<br>

– сортировка слов по размеру;<br>
– сортировка слов в лексикографическом порядке.<br>
Примечание. Разделители слов можно задавать отдельно. По 
умолчанию в качестве разделителя принимается только символ 
пробела.

```python
class StringFormatter:

    @staticmethod
    def deleteLess(text, n):
        list = text.split()
        list = [item for item in list if len(item) >= n]
        return " ".join(list)

    @staticmethod
    def replaceNum(text):
        for word in text:
            if word.isdigit():
                text = text.replace (word, "*")
        return text

    @staticmethod
    def insertSpace(text):
        return " ".join(text)
   
    @staticmethod
    def sortBySize(text):
       return " ".join(sorted(text.split(), key=len))
  
    @staticmethod
    def sortByAlphabet(text):
        return " ".join(sorted(text.split()))

print(StringFormatter.deleteLess("abcdefg abc abbbbbb abc", 5))
print(StringFormatter.replaceNum("abb023 2021 2022 2abd3"))
print(StringFormatter.insertSpace("insert space"))
print(StringFormatter.sortBySize("sort by size"))
print(StringFormatter.sortByAlphabet("sort by alphabet"))
        
```

Примеры окон приведены на рис. 4 (все элементы управления необходимо обязательно реализовать те же, что присутствуют на рисунке). Разные комбинации отмеченных чекбоксов приводят к разным цепочкам операций форматирования задаваемой в верхнем поле строки с разными результатами.

```python
import wx

class StringFormatter:

    @staticmethod
    def deleteLess(text, n):
        list = text.split()
        list = [item for item in list if len(item) >= n]
        return " ".join(list)

    @staticmethod
    def replaceNum(text):
        for word in text:
            if word.isdigit():
                text = text.replace (word, "*")
        return text

    @staticmethod
    def insertSpace(text):
        return " ".join(text)
   
    @staticmethod
    def sortBySize(text):
       return " ".join(sorted(text.split(), key=len))
  
    @staticmethod
    def sortByAlphabet(text):
        return " ".join(sorted(text.split()))

class StringFormatterGUI(wx.Frame):

    def __init__(self, parent = None):
        super().__init__(parent)
        self.SetTitle('StringFormatter Demo')
        self.Center()
        self.SetSize((400, 320))
        path = os.getcwd()
        ico = wx.Icon( path + '\\' + 'string.ico', wx.BITMAP_TYPE_ICO)
        self.SetIcon(ico)
        self.SetMaxSize((400, 320))
        self.SetMinSize((400, 320))
        panel = wx.Panel(self, -1)
        wx.StaticText(panel, -1, "Строка:", pos=(5, 20))
        self.enterText = wx.TextCtrl(panel, -1, "", size=(300, -1), pos=(70, 20))
        self.resultLabel = wx.StaticText(panel, -1, "Результат:", pos=(5, 230))
        self.resultText = wx.TextCtrl(panel, -1, "", size=(300, -1), pos=(70, 230))
        self.lessSpinCtrl = wx.SpinCtrl(panel, -1, "", (292, 60), (40, -1), min=1, max=20)
        self.dellLessCheckBox = wx.CheckBox(panel, -1, "Удалить слова размером меньше",(70, 60), (210, 20))
        self.replaceNumCheckBox = wx.CheckBox(panel, -1, "Заменить все цифры на *", (70, 80), (220, 20))
        self.insertSpaceCheckBox = wx.CheckBox(panel, -1, "Вставить пробелы между символами", (70, 100), (280, 20))
        self.sortCheckBox = wx.CheckBox(panel, -1, "Сортировать слова в строке", (70, 120), (220, 20))
        self.sortBySizeRadioButton = wx.RadioButton(panel, -1, "По размеру", (100, 140), (150, 20))
        self.sortByAlphabetRadioButton = wx.RadioButton(panel, -1, "По алфавиту", (100, 160), (150, 20))
        self.sortBySizeRadioButton.Disable()
        self.sortByAlphabetRadioButton.Disable()
        self.button = wx.Button(panel, -1, "Форматировать", size=(300, 30), pos=(70, 190))
        self.Bind(wx.EVT_BUTTON, self.buttonCick, self.button)
        self.Bind(wx.EVT_CHECKBOX, self.sortCheckBoxChecked, self.sortCheckBox)
        

    def buttonCick(self, event):
        sf = StringFormatter()
        result = self.enterText.Value
        if self.dellLessCheckBox.IsChecked():
            result = sf.deleteLess(result, self.lessSpinCtrl.Value)
        if self.replaceNumCheckBox.IsChecked():
            result = sf.replaceNum(result)
        if self.insertSpaceCheckBox.IsChecked():
            result = sf.insertSpace(result)
        if self.sortCheckBox.IsChecked():
            if self.sortBySizeRadioButton.GetValue():
                result = sf.sortBySize(result)
            elif self.sortByAlphabetRadioButton.GetValue():
                result = sf.sortByAlphabet(result)
        self.resultText.SetValue(result)

    def sortCheckBoxChecked(self, event):
        if self.sortCheckBox.IsChecked():
            self.sortBySizeRadioButton.Enable()
            self.sortByAlphabetRadioButton.Enable()
        else:
            self.sortBySizeRadioButton.Disable()
            self.sortByAlphabetRadioButton.Disable()


if __name__ == '__main__':

    app = wx.App()
    sf = StringFormatterGUI()
    sf.Show()
    app.MainLoop()
```
